### Read mvps CSV

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

### Remove redundant data from mvps df

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps
```

### Read players CSV

```python
players = pd.read_csv("players.csv")
```

```python
players
```

### Remove redundant data from players df

```python
del players["Unnamed: 0"]
```

```python
del players["Rk"]
```

### Remove asterisks (*) from any player names

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991))
```

### Get only total points for each player

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1, :]["Tm"]
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

### Flatten indices

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

### Merge mvps and players dfs

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head(5)
```

```python
combined[combined["Pts Won"] > 0]
```

### Replace missing values with 0

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined[combined["Pts Won"] > 0]
```

### Load teams CSV

```python
teams = pd.read_csv("teams.csv")
```

```python
teams.head(30)
```

### Remove Division rows

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(30)
```

### Remove asterisks (*) from any team names

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams.head(5)
```

### Load mapping for team names and abbreviations

```python
abbreviations = {}

with open("abbreviations.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        abbreviations[abbrev] = name
```

### Map team names to combined df

```python
combined["Team"] = combined["Tm"].map(abbreviations)
```

```python
combined.head(5)
```

### Merge combined and teams df

```python
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```

```python
stats
```

```python
combined
```

### Remove Unnamed column

```python
del stats["Unnamed: 0"]
```

```python
stats.dtypes
```

### Transform to numeric where applicable

```python
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```

```python
stats["GB"].unique()
```

### Replace dash with 0 in GB

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
```

```python
stats["GB"].unique()
```

### Transform GB to numeric

```python
stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats.dtypes
```

### Save to CSV

```python
stats.to_csv("player_mvp_stats.csv")
```
